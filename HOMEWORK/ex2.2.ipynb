```python
from matplotlib import pyplot as plt
import numpy as np
x = np.linspace(-2, 2, 100)
sum = 0
for k in range(1,100):
    a_k = -4 * np.cos(np.pi * k / 2) / (np.pi * k)**2
    b_k = 0
    a_k_mode = a_k * np.cos(np.pi * k * x / 2)
    b_k_mode = b_k * np.sin(np.pi * k * x / 2)
    sum = sum + a_k_mode + b_k_mode
plt.plot(x, sum)
```
