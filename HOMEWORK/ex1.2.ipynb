```python
import os
from matplotlib.image import imread
import numpy as np

image = imread(os.path.join("..", "DATA", "dog.jpg"))
image = np.mean(image, -1)
n, m = image.shape

errors = []
rangeR = range(1, n + 1)
U, S, VT = np.linalg.svd(image, full_matrices=False)

for r in rangeR:
    print(f"r = {r}")
    reconstructed_image = U[:, :r] @ np.diag(S)[:r, :r] @ VT[:r, :]
    errors.append(np.linalg.norm(reconstructed_image - image, ord="fro"))
```

```python
import matplotlib.pyplot as plt
plt.plot(rangeR, errors)
plt.xlabel("r")
plt.ylabel("Error")
plt.title("Error vs r")
```

```python
squared_errors = np.array(errors) ** 2
total_error = np.sum(squared_errors)
missing_error_ratios = []
for r in rangeR:
    explained_error = np.sum(squared_errors[:r])
    missing_error = total_error - explained_error
    missing_error_ratios.append((missing_error / total_error))

plt.plot(rangeR, missing_error_ratios)
plt.xlabel("r")
plt.ylabel("Missing error ratio")
plt.show()
captured_norm_ratios = 1 - np.array(missing_error_ratios)
plt.plot(rangeR, captured_norm_ratios)
plt.xlabel("r")
plt.ylabel("Captured norm ratio")
plt.show()
```

```python
plt.plot(np.cumsum(S) / np.sum(S))
plt.xlabel("r")
plt.ylabel("Cumulative sum of singular values")
plt.show()
plt.plot(np.cumsum(S**2) / np.sum(S**2))
plt.xlabel("r")
plt.ylabel("Cumulative sum of the variance")
plt.show()
```

```python
threshold = 0.99

index = np.argmax(captured_norm_ratios >= threshold)
first_value = rangeR[index]

print(first_value)
```

```python
index = np.argmax(np.cumsum(S) / np.sum(S) >= threshold)
first_value = rangeR[index]

print(first_value)
```
