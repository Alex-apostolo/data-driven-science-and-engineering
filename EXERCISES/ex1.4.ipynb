```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)

for n in [50, 100, 200, 500, 1000]:
    S_array = []
    for i in range(100):
        X = np.random.normal(0, 1, (n, n))
        U, S, VT = np.linalg.svd(X)
        S_array.append(S)

    # Create the subplots using plt.subplots
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    fig.suptitle(f'n = {n}')

    # First subplot: Boxplot of all singular values
    axs[0].boxplot(np.concatenate(S_array))
    axs[0].set_ylabel('Singular values')
    axs[0].set_title('Boxplot of Singular Values')

    # Second subplot: Mean singular values
    axs[1].plot(np.mean(S_array, axis=1), marker='o')
    axs[1].set_ylabel('Mean singular values')
    axs[1].set_xlabel('r')
    axs[1].set_title('Mean Singular Values')

    # Third subplot: Median singular values
    axs[2].plot(np.median(S_array, axis=1), marker='o')
    axs[2].set_ylabel('Median singular values')
    axs[2].set_xlabel('r')
    axs[2].set_title('Median Singular Values')

    # Adjust layout to prevent overlap
    fig.tight_layout()

    # Show the plot
    plt.show()
```
