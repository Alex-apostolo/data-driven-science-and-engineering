```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
dims = np.arange(2, 11, 1)
fraction = np.zeros(len(dims))
for N in dims:
    scaled_array = np.random.rand(10000, N) * 2
    circle_center = np.ones(N)
    distances = np.linalg.norm(scaled_array - circle_center, axis=1)
    inside_circle = np.sum(distances < 1)
    fraction[N - 2] = inside_circle / 10000
    plt.hist(distances, bins=30, alpha=0.5, label='N = {}'.format(N))

plt.show()
plt.plot(dims, fraction, 'o-')
plt.xlabel('Dimensions')
plt.ylabel('Fraction of points inside the unit circle')
plt.title('Fraction of points inside the unit circle in different dimensions')
```
