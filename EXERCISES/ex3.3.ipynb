```python
import numpy as np
np.random.seed(0)
K = 5
```

```python
from cosamp import cosamp
import scipy


all_results = []
for n in [100, 1000, 2000, 5000]:
    Phi = scipy.fft.dct(np.eye(n), axis=0, norm="ortho")
    t = np.linspace(0, 1, n)
    signal = np.cos(2 * np.pi * 100 * t) + np.cos(2 * np.pi * 900 * t)
    random_indices = np.random.choice(n, K, replace=False)
    s = np.zeros(n)
    s[random_indices] = signal[random_indices]

    results = []
    for i in range(1, 101):
        C = np.random.rand(i, n)
        y = C @ Phi @ s
        Theta = C @ Phi
        results.append(cosamp.cosamp(Theta, y, K, 1e-10, 100))
    all_results.append(results)
```

```python
import matplotlib.pyplot as plt

def plot_errors():
    all_errors = []
    for i, results in enumerate(all_results):
        errors = [np.linalg.norm(s - result) / np.linalg.norm(s) for result in results]
        all_errors.append(errors)
    plt.boxplot(all_errors, showfliers=False)
    plt.show()
plot_errors()
```
